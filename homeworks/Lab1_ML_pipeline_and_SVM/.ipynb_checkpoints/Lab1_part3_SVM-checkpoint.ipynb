```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
#    'kernel' :  ['poly', 'rbf', 'sigmoid'],
    'C' : range(1, 10),
    'gamma' : ['scale', 'auto'], 
    'degree' : range(1, 6),
    'coef0' : range(1, 10, 2)
}


for kernel in ['poly', 'rbf','sigmoid']:
    grid_search = GridSearchCV(SVC(kernel = kernel), param_grid, scoring = 'f1', cv = 3)
    grid_search.fit(moons_points, moons_labels)
    print(f'f1_score : {grid_search.best_score_}')
    print(f'params : {grid_search.best_params_}')


    plot_decisions(moons_points, moons_labels, grid_search.best_estimator_, f'SVM with {kernel} kernel')
```

```python
def plot_decisions_of_grid(grid, kernel):
    clfs = []
    clfs_names = []
    for C in grid['C']:
        for gamma in grid['gamma']:
            for degree in grid['degree']:
                clf = SVC(kernel = kernel, C = C, degree = degree, gamma = gamma)
                clfs.append(clf)
                clfs_names.append(f'SVC(kernel = {kernel}, C = {C}, degree = {degree}, gamma = {gamma})')
    for clf, clf_name in zip(clfs, clfs_names):
        clf.fit(moons_points, moons_labels)
        plot_decisions(moons_points, moons_labels, clf, clf_name)
```

```python
for degree in range(1, 10):
        lr = SVC(kernel = 'poly', C = 6, gamma = 'auto', degree = degree, coef0 = 6)
        lr.fit(moons_points, moons_labels)
        plot_decisions(moons_points, moons_labels, lr, f'SVC(kernel = poly, C = 6, gamma = auto, degree = {degree})')
```

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

```

```python
X_train, X_test, y_train, y_test = train_test_split( moons_points, moons_labels, test_size=0.3, random_state=42)
for degree in range(1, 12):
    pline = Pipeline([
    ('pf', PolynomialFeatures(degree = degree)),
    ('lr', LogisticRegression())])

    pline.fit(X_train, y_train)
    print(f'f1_score: {f1_score(y_test, pline.predict(X_test))}')
    plot_decisions(moons_points, moons_labels, pline, f'Logistic Regression with polinomial features degree {degree}')
```

```python
pline = Pipeline([
    ('pf', PolynomialFeatures(degree = 4)),
    ('lr', LogisticRegression())])

pline.fit(X_train, y_train)
print(f'f1_score: {f1_score(y_test, pline.predict(X_test))}')
plot_decisions(moons_points, moons_labels, pline, f'Logistic Regression with polinomial features degree 4')
```

```python
from sklearn.datasets import make_circles

circles_points, circles_labels = make_circles(n_samples=500, noise=0.06, random_state=42)

plt.figure(figsize=(5, 5))
plt.scatter(circles_points[:, 0], circles_points[:, 1], c=circles_labels)
```

```python
param_grid = {
#    'kernel' :  ['poly', 'rbf', 'sigmoid'],
    'C' : np.logspace(-2,1,10),
    'gamma' : ['scale', 'auto'], 
    'degree' : range(1, 6),
    'coef0' : range(1, 10, 2)
}


for kernel in ['poly', 'rbf','sigmoid']:
    grid_search = GridSearchCV(SVC(kernel = kernel), param_grid, scoring = 'f1', cv = 3)
    grid_search.fit(circles_points, circles_labels)
    print(f'f1_score : {grid_search.best_score_}')
    print(f'params : {grid_search.best_params_}')


    plot_decisions(circles_points, circles_labels, grid_search.best_estimator_, f'SVM with {kernel} kernel')
```

And even more:

```python
points = np.vstack((circles_points*2.5 + 0.5, moons_points))
labels = np.hstack((circles_labels, moons_labels + 2)) # + 2 to distinct moons classes

plt.figure(figsize=(5, 5))
plt.scatter(points[:, 0], points[:, 1], c=labels)
```

```python
param_grid = {
    'C' : np.logspace(-2,1,10),
    'gamma' : ['scale', 'auto'], 
    'degree' : range(1, 6),
    'coef0' : range(1, 10, 2)
}


for kernel in ['poly', 'rbf']:
    grid_search = GridSearchCV(SVC(kernel = kernel), param_grid, scoring = 'f1_macro', cv = 3)
    grid_search.fit(points, labels)
    print(f'f1_score : {grid_search.best_score_}')
    print(f'params : {grid_search.best_params_}')


    plot_decisions(points, labels, grid_search.best_estimator_, f'SVM with {kernel} kernel')
```
